```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def fun(x):
    offset_rear = -1
    offset_front = 1
    mu_rear = np.array([-1,0])
    mu_front = np.array([1,0])
    sigmainv_rear = np.array([[1,0],[0,1]])
    sigmainv_front = np.array([[1,0],[0,1]])
    
    #front gauss value
    front = np.exp(-(x-mu_rear)@sigmainv_rear@(x-mu_rear))
    back = np.exp(-(x-mu_front)@sigmainv_front@(x-mu_front))
    return front - back
```

```python

nx = 30
ny = 30
x = np.linspace(-6, 6, nx)
y = np.linspace(-6, 6, ny)
X, Y = np.meshgrid(x, y)
xre = X.reshape(-1,)
yre = Y.reshape(-1,)

Zvals = np.zeros((nx*ny,))
for idx in range(nx*ny):
    Zvals[idx]= fun(np.array([xre[idx], yre[idx]]))
Z = Zvals.reshape(nx,ny)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
